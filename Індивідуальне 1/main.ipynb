```python
import matplotlib.pyplot as plt
from sympy.solvers import solve
from sympy import Symbol
import numpy as np
```

```python
A = [
    [10, 3],
    [-1, 1],
    [1, 1],
    [0, 1],
]
```

```python
b = [
    30, 
    5,
    10,
    2,
]
```

```python
sig = [
    '>=',
    '<=',
    '<=',
    '>=',
]
```

```python
# draw the inequalities
x2 = Symbol('x2')
lambdas = []
for i in range(len(A)):
    if A[i][1] != 0:
        lambdas.append(lambda x2: (b[i] - A[i][0] * x2) / A[i][1])
    else:
        lambdas.append(lambda x2: b[i] - A[i][0] * x2)


x2 = np.linspace(-10, 10, 1000)
for i in range(len(A)):
    y = lambdas[i](x2)
    if sig[i] == '<=':
        plt.plot(x2, y, label=f'{A[i][0]}x + {A[i][1]}y <= {b[i]}', color='blue')
        plt.fill_between(x2, y, np.full((len(x2)), 10), color='blue', where=y <= 10, alpha=0.3)#
    else:
        plt.plot(x2, y, label=f'{A[i][0]}x + {A[i][1]}y >= {b[i]}', color='red')
        plt.fill_between(x2, y, np.full((len(x2)), -10), color='red', where=y >= -10, alpha=0.3)#

plt.xlabel('x1')
plt.ylabel('x2')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
```
